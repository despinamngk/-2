```python
from bs4 import BeautifulSoup
import os
import requests
from datetime import datetime
import ast
import time


list=[]
for i in urls:  
    dic = {}
    print(i)
    page = requests.get(i, headers = headers)
    soup = BeautifulSoup(page.content, 'html.parser')

    
    title = soup.select("h1")

    if (title):
        dic['title'] = soup.select("title")[0].text
    else:
        print("ARTICLE NO: ", link, " HAS NO TITLE")
        dic['title'] = "NO TITLE"
    
    Date = soup.select("p time")

    if (Date):
        dic['Date'] = soup.select("p time")[0].text.strip()
    else:
        print("ARTICLE NO: ", link, " HAS NO date")
        dic['Date'] = "NO date"
  
    
    article = soup.select("div.main-content.story-content")
    if (article):
        dic['article'] =  article = soup.select("div.main-content.story-content")[0].text
    else:
        print("ARTICLE NO: ", link, " HAS NO article")
        dic['article'] = "NO ARTICLE"  

    category= soup.select("a.main-item-category")
    if (category):
        dic['category'] =  soup.select("a.main-item-category")[0].text.strip()
    else:
        print("ARTICLE NO: ", link, " HAS NO category")
        dic['category'] = "NO category"  
  
    
    dic['url'] = i

    
    list.append(dic)
    #print(dic)
    
    time.sleep(0.5) 
```
